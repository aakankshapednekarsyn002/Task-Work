```python
# Feature Relationships
# Scatter plot RAM vs Price
plt.figure(figsize=(10, 8))
plt.scatter(df['ram'], df['price_euros'])
plt.title("RAM vs Laptop Price")
plt.xlabel("RAM")
plt.ylabel("Price (Euros)")
plt.tight_layout()
plt.show()
```

```python
avg_price = df.groupby('company')['price_euros'].mean().sort_values()
print(avg_price)
```

```python
# Average Laptop Price by Company
plt.figure(figsize = (12,7))
avg_price.plot(kind="bar", color = "DarkBlue")
plt.title("Average Laptop Price vs Company")
plt.xlabel("Company")
plt.ylabel("Average Price(Euros)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Histogram for Price Distribution
plt.figure(figsize = (12,7))
plt.hist(df['price_euros'], bins = 40)
plt.title("Distribution of Laptop Prices")
plt.xlabel("Price(Euros)")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#changed datatypes for heatmap for required columns
df['ram'] = df['ram'].astype(str).str.replace('GB','').astype(int)
df['weight'] = df['weight'].astype(str).str.replace('kg','').astype(float)
df['laptop_id'] = df['laptop_id'].astype('object')

print(df.dtypes)
```

```python
# Numeric Columns for Heatmap
numeric_df = df.select_dtypes(include=['int64','float64', 'int32'])
correlation_matrix = numeric_df.corr()

# Heatmap
plt.figure(figsize=(12,7))
plt.imshow(correlation_matrix, cmap='coolwarm' , interpolation='nearest')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Box plot for Laptop Prices
plt.figure(figsize=(12,7))
plt.boxplot(df['price_euros'])
plt.title("Box Plot For Laptop Prices")
plt.ylabel("Price (Euros)")
plt.tight_layout()
plt.show()
```

```python
# IQR (Interquartile Range)
Q1 = df['price_euros'].quantile(0.25)
Q3 = df['price_euros'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Outliers using IQR
outliers = df[(df['price_euros'] < lower_bound) | (df['price_euros'] > upper_bound)]


print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers using IQR:", outliers.shape[0])
```

```python
plt.figure(figsize=(12,7))
plt.scatter(df.index, df['price_euros'])
plt.scatter(outliers.index, outliers['price_euros'])
plt.title("Outliers using IQR method (Price)")
plt.xlabel("Index")
plt.ylabel("Price(Euros)")
plt.show()
```

```python
# Z score Analysis
mean = df['price_euros'].mean()
std = df['price_euros'].std()

# Z-score 
z_score = (df['price_euros'] - mean) / std
outliers_zcore = df[np.abs(z_score) > 3]

print("Number of Outliers:", len(outliers_zscore))
```
